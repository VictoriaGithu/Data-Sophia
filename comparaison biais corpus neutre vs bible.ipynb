```python
# Diviser le texte en lignes
lignes_bible = texte_avec_bible.strip().split('\n')

# Initialiser la liste de listes
liste_de_listes_bible = []

# Parcourir chaque ligne
for ligne in lignes_bible:
     # Vérifier si la ligne contient des accolades
    if '{' in ligne and '}' in ligne:
        # Extraire les mots entre les accolades
        mots = ligne.split("{")[1].split("}")[0].replace("'", "").split(", ")
        # Ajouter la liste de mots à la liste de listes
        liste_de_listes_bible.append(mots)

# Afficher le résultat
print(liste_de_listes_bible)
```

```python
# Initialiser le dictionnaire pour stocker les occurrences
occurrences = {}

# Calculer les occurrences de chaque mot
for liste_mots in liste_de_listes_bible:
    for mot in liste_mots:
        occurrences[mot] = occurrences.get(mot, 0) + 1

# Trier les mots par nombre d'occurrences
occurrences = dict(sorted(occurrences.items(), key=lambda item: item[1], reverse=True))

# Extraire les mots et leurs occurrences
mots = list(occurrences.keys())
occurrences = list(occurrences.values())

mots_rouge = ['veuve', 'princesse', 'semme', 'princesse']
mots_vert = ['servante', 'soeur', 'mere', 'mari', 'concubine', 'epouse']
mots_bleu = ['nourrice', 'fille']

# Créer une liste de couleurs pour les barres
couleurs = ['black' if mot in mots_rouge else 'green' if mot in mots_vert else 'violet' if mot in mots_bleu else 'gray' for mot in mots]

# Créer l'histogramme
plt.figure(figsize=(10, 6))
plt.bar(range(len(mots)), occurrences, align='center', color = couleurs)
plt.xticks(range(len(mots)), mots, rotation=45, ha='right')
plt.xlabel('Mots')
plt.ylabel('Occurrences')
plt.title('Occurrences des mots dans corpus neutre + BIBLE')
#plt.yticks(range(1, 11))
plt.tight_layout()
plt.show()
```

```python
# Diviser le texte en lignes
lignes_neutre = text_neutre.strip().split('\n')

# Initialiser la liste de listes
liste_de_listes_neutre = []

# Parcourir chaque ligne
for ligne in lignes_neutre:
     # Vérifier si la ligne contient des accolades
    if '{' in ligne and '}' in ligne:
        # Extraire les mots entre les accolades
        mots = ligne.split("{")[1].split("}")[0].replace("'", "").split(", ")
        # Ajouter la liste de mots à la liste de listes
        liste_de_listes_neutre.append(mots)

# Afficher le résultat
print(liste_de_listes_neutre)
```

```python
# Initialiser le dictionnaire pour stocker les occurrences
occurrences = {}

# Calculer les occurrences de chaque mot
for liste_mots in liste_de_listes_neutre:
    for mot in liste_mots:
        occurrences[mot] = occurrences.get(mot, 0) + 1

# Trier les mots par nombre d'occurrences
occurrences = dict(sorted(occurrences.items(), key=lambda item: item[1], reverse=True))

# Extraire les mots et leurs occurrences
mots = list(occurrences.keys())
occurrences = list(occurrences.values())

# Créer l'histogramme
plt.figure(figsize=(10, 6))
plt.bar(range(len(mots)), occurrences, align='center', color='gray')
plt.xticks(range(len(mots)), mots, rotation=45, ha='right')
plt.xlabel('Mots')
plt.ylabel('Occurrences')
plt.title('Occurrences des mots corpus neutre')
plt.tight_layout()
plt.show()
```
